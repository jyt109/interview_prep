2. **Explain the difference between mutually exclusive and independence.**

   - Mutually exclusive means the occurrence of one event precludes the others
   - Independence means there is no relationship between the occurrences of the events
   - Hence mutually exclusive events are not independent and vice versa

4. **Compute the variance of the following population: `10, 12, 34, 24`**

   - $\bar{X} = \frac{10+12+34+24}{4} = 20$
   - $Var(X) = \frac{(10-20)^2 + (12-20)^2 + (34-20)^2 + (24-20)^2}{4} = 94$
   
   <br>

5. **How would you prove if a coin is unfair ?**

   - Let $X$ be the number of heads when flip 1000 times
     $$X \sim Binomial(n=1000, p=0.5)$$
   - The mean ($mu$) is $1000(0.5) = 500$
   - The standard deviation ($sigma$) is $\sqrt{1000(0.5)(0.5)} = \sqrt{250}$
   
     <br>
     
   - By CLT:
     $$X \sim Normal(\mu=500, \sigma=\sqrt{250})$$
     
     <br>
     
   - Flipped coin 1000 times and got head $H$ number of times
   - Compute Z-score for $H$:
     $$Z = \frac{H - 500}{\sqrt{250}}$$
     
     <br>
     
   - Compute p-value of observing $H$:
     $$p = 1 - cdf(Z)$$
     
     <br>
     
     If $p < 0.05 /2$ (two-tailed), then null ($\mu$=500) is rejected and coin is unfair
     
   <br>
   
6. **Explain what the standard error of mean is.** 

   - Standard deviation of the sampling distribution of the mean 
   - Given we have estimated the population mean ($\mu$) from a sample mean ($\bar{X}$)
   - There is a distribution of sample means we could have observed given a certain population
   - We can computer the standard error as such:
    
     $$SE = \frac{\text{sample SD}}{\sqrt{\text{sample size}}} $$
   - The smaller the sample SD or the larger the sample size, the smaller the standard error is

   <br>
   
7. **What is a confidence interval and what does it represents ?**

   - A 95% confidence interval means we are 95% confident that the interval include the population statistics
   - Meaning if 100 samples are drawn and 100 confidence intervals are computed, 95 will contain the population statistic and 5 would not
  
   <br>

8. **How do you calculate the 95% confidence interval of a sample mean ?**

   - Given a large enough sample size (>30), the sampling distribution of the mean will converge to a normal distribution (Central Limit Theorem)
   - 95% CI translates to $\pm 1.96 \times \text{standard error of sample mean}$

   <br>

9. **Describe the 2 ways to decide the parameters for the given distribution that best fit the dataset ?**

   - **Method of Moment (MOM):**
   
     - Estimate the population statistics from the sample statistics (mean, variance...) and computer paramters for distribution
     
     <br>
     
   - **Maximum Likelihood Estimate (MLE):**
     
     - Compute the likelihood of observing all the data points given certain distribution paramters (product of the PMF / PDF of all the data points) 
     - Select the paramters that yields the highest likelihood

   <br>

10. **Describe the central limit theorem. As a data scientist, state a few scenarios where you cannot use the CLT.**

    - The arithmetic mean of a sufficiently large number of iterates of independent random variables, each with a well-defined expected value and well-defined variance, will be approximately normally distributed, regardless of the underlying distribution
    - **Cannot use CLT if:**
      - Sampling is not random
      - Underlying distribution does not have define mean / variance (power law distribution)
      - The statistic is not arithmetic mean (median / variance ...)

   <br>

11. **Describe what statistical power is. Illustrate the concept with an example.**

    - The ability to reject the null hypothesis given the alternative hypothesis is true
    - i.e. Probability of false negative
    - The test is not _powerful_ enough if the effect size we are trying to detect is small, i.e. the alternative hypothesis is very similar to the null
    - Or the sample size is small, i.e. we do not have a large enough sample size (too little evidence) to reject the null

   <br>
   
12. **What is the difference of using the mean vs the median as a centrality measure ?**

    - Median does not account for the values at the tails (smallest and largest) values of the distribution 
    - Hence median is more insensitive to extreme values / outliers
    - Often times the arithmetic mean does not sufficiently inform us of the distribution of data
    - 25%, 50% (median), 75% quantiles, variance and skewness can inform us more about the distribution of values

   <br>
